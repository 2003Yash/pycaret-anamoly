```python
import pandas as pd
df = pd.read_csv('/content/creditcarddata.csv')
df.head()
```

```python
!pip install pycaret  # Install the main PyCaret package
from pycaret.anomaly import setup, create_model, assign_model, evaluate_model, predict_model, save_model, load_model  # Import from the correct module
```

```python
#perform all necessary pre-processing automatically using setup and create a dataframe
s = setup(df)
```

```python
my_model = create_model('knn')
```

```python
assign_model(my_model)
```

```python
evaluate_model(my_model)
```

```python
predict_model(my_model, data = df)
```

```python
save_model(my_model, 'knn')
```
